```python
import pandas as pd
train_data_file = "/home/zxp/learn/datasets/2015数据集/twitter2015_gpt_all/train.tsv"
train_data = pd.read_csv(train_data_file, sep='\t',)

test_data_file = "/home/zxp/learn/datasets/2015数据集/twitter2015_gpt_all/test.tsv"
test_data = pd.read_csv(test_data_file, sep='\t',)

dev_data_file = "/home/zxp/learn/datasets/2015数据集/twitter2015_gpt_all/dev.tsv"
dev_data = pd.read_csv(dev_data_file, sep='\t',)

# 拼接一块
all_data = pd.concat([train_data, test_data, dev_data], ignore_index=True)

#保存为excel
all_data_file = "/home/zxp/learn/datasets/2015数据集/twitter2015_gpt_all/all_data.xlsx"
all_data.to_excel(all_data_file, index=False)
```
